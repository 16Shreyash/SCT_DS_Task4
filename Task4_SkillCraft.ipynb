```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as st
import seaborn as sns
sns.set(style='darkgrid')
```

```python
data = pd.read_csv(r'C:\Users\shrey\OneDrive\Desktop\RTA Dataset.csv')
```

```python
data['Time'] = pd.to_datetime(data['Time'], errors='coerce').dt.hour
key_columns = ['Time', 'Day_of_week', 'Road_surface_type', 'Road_surface_conditions', 
               'Light_conditions', 'Weather_conditions', 'Accident_severity']
data_cleaned = data.dropna(subset=key_columns)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data_cleaned['Time'], bins=24, kde=True, color='skyblue')
plt.title('Accidents by Time of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Accident Count')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=data_cleaned, x='Road_surface_conditions', hue='Accident_severity', palette='viridis')
plt.title('Accident Severity by Road Surface Conditions')
plt.xlabel('Road Surface Conditions')
plt.ylabel('Accident Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=data_cleaned, x='Weather_conditions', hue='Accident_severity', palette='magma')
plt.title('Accident Severity by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Accident Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data_cleaned, x='Day_of_week', 
              order=["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"], palette="coolwarm",hue='Day_of_week',legend=False)
plt.title('Accident Counts by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Accident Count')
plt.grid(True)
plt.show()
```
